```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# JSON dosyasının yolu
file_path = r"C:\Users\Muhammed\OneDrive\Masaüstü\emsal.json"

# JSON dosyasını DataFrame olarak okuma
df = pd.read_json(file_path, encoding="utf-8")

# İlk birkaç satırı görüntüleme
print(df.head())
```

```python
df.head()
```

```python
# Boşluk ve özel karakterleri temizle
df['content'] = df['content'].str.replace(r'\n', ' ', regex=True).str.strip()
```

```python
df
```

```python
df["content"] = df["content"].str.strip() #başındaki ve sonundaki boşlukların silinmesi
```

```python
df.drop_duplicates(inplace=True) #tekrar eden satırların  kaldırılması
```

```python
#Veri Ön İzleme
def data(dataframe,head=True):
    print("----------------------------SHAPE------------------------------------")
    print(dataframe.shape)
    print("----------------------------İNFO-------------------------------------")
    print(dataframe.info())
    print("----------------------------NUNİQUE-----------------------------------")
    print(dataframe.nunique())
    print("----------------------------İSNULL------------------------------------")
    print(dataframe.isnull().sum().sort_values(ascending=False))
    print("----------------------------DESCRİBE----------------------------------")
    print(dataframe.describe())
    if head:
        print("-----------------------HEAD---------------------------------------")
        print(dataframe.head())
        print("-----------------------TAİL---------------------------------------")
        print(dataframe.tail())
```

```python
data(df,head=True)
```

```python
# Şehir isimlerinin hem büyük hem küçük harfli versiyonlarını içeren liste
sehirler = [
    'ADANA', 'ADIYAMAN', 'AFYONKARAHİSAR', 'AĞRI', 'AKSARAY', 'AMASYA', 'ANKARA', 'ANTALYA', 'ARDAHAN', 'ARTVİN', 
    'AYDIN', 'BALIKESİR', 'BARTIN', 'BATMAN', 'BAYBURT', 'BİLECİK', 'BİNGÖL', 'BİTLİS', 'BOLU', 'BURDUR', 'BURSA', 
    'ÇANAKKALE', 'ÇANKIRI', 'ÇORUM', 'DENİZLİ', 'DİYARBAKIR', 'DÜZCE', 'EDİRNE', 'ELAZIĞ', 'ERZİNCAN', 'ERZURUM', 
    'ESKİŞEHİR', 'GAZİANTEP', 'GİRESUN', 'GÜMÜŞHANE', 'HAKKARİ', 'HATAY', 'IĞDIR', 'ISPARTA', 'İSTANBUL', 'İZMİR', 
    'KAHRAMANMARAŞ', 'KARABÜK', 'KARAMAN', 'KARS', 'KASTAMONU', 'KAYSERİ', 'KIRIKKALE', 'KIRKLARELİ', 'KIRŞEHİR', 
    'KİLİS', 'KOCAELİ', 'KONYA', 'KÜTAHYA', 'MALATYA', 'MANİSA', 'MARDİN', 'MERSİN', 'MUĞLA', 'MUŞ', 'NEVŞEHİR', 
    'NİĞDE', 'ORDU', 'OSMANİYE', 'RİZE', 'SAKARYA', 'SAMSUN', 'SİİRT', 'SİNOP', 'SİVAS', 'ŞANLIURFA', 'ŞIRNAK', 
    'TEKİRDAĞ', 'TOKAT', 'TRABZON', 'TUNCELİ', 'UŞAK', 'VAN', 'YALOVA', 'YOZGAT', 'ZONGULDAK'
]

# Küçük harfli versiyonlarını da ekleyelim
sehirler_lower = [sehir.lower() for sehir in sehirler]

# Şehirleri hem büyük hem küçük harfli olarak eşleştirecek fonksiyon
def sehir_bul(content):
    content_lower = content.lower()  # İçeriği küçük harfe çevir
    for buyuk, kucuk in zip(sehirler, sehirler_lower):
        if buyuk in content or kucuk in content_lower:
            return buyuk  # Büyük harfli halini döndür
    return 'Diğer'  # Hiçbiri bulunmazsa 'Diğer'

# DataFrame'e yeni bir sütun ekleyelim
df['Sehir'] = df['content'].apply(sehir_bul)

# Sonucu kontrol edelim
print(df[['content', 'Sehir']].head(20))

```

```python
df.head()
```

```python
# En çok geçen 10 şehir
top_cities = df['Sehir'].value_counts().nlargest(10)

# En az geçen 10 şehir
least_cities = df['Sehir'].value_counts().nsmallest(10)

# Grafik boyutu ayarla
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# En çok geçen 10 şehir grafiği
sns.barplot(x=top_cities.index, y=top_cities.values, palette="viridis", ax=axes[0])
axes[0].set_title("En Çok Geçen İlk 10 Şehir", fontsize=14)
axes[0].set_xlabel("Şehir", fontsize=12)
axes[0].set_ylabel("Adet", fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

# En az geçen 10 şehir grafiği
sns.barplot(x=least_cities.index, y=least_cities.values, palette="coolwarm", ax=axes[1])
axes[1].set_title("En Az Geçen 10 Şehir", fontsize=14)
axes[1].set_xlabel("Şehir", fontsize=12)
axes[1].set_ylabel("Adet", fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

# Grafikleri göster
plt.tight_layout()
plt.show()
```
